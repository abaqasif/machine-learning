```python
import imageio as img #for reading the images
import numpy as np #for ndarray
import pandas as pd #for dataframe
import os #for manipulation with files in the directory
from sklearn import manifold #to run isomap
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as Axes3D
%matplotlib notebook
```

```python
#change working directory to where the image exists
os.chdir("C:\\D drive\Data Science with Python\\DS with Python\\Microsoft-DAT210x-master\\Microsoft-DAT210x-master\\Module-4\\Datasets\\ALOI\\32")
#get list of all image files present in the directory
list_of_image_files=os.listdir()
#print all the image files that exist in the directory
print(list_of_image_files)
images=[]
```

```python
#iterate to store all images in a list if ndarrays
for i in range(len(list_of_image_files)):
    images.insert(i,np.array(img.imread(list_of_image_files[i]).reshape(-1)))
```

```python
#pixel size of each image is 144 x 192
print(images[0].shape)
```

```python
df=pd.DataFrame(images)
```

```python
# Run Isomap on the DataFrame:
iso = manifold.Isomap(n_neighbors = 6, n_components = 3)
T = iso.fit_transform(df)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.set_title("3 components ISOMAP")
ax.set_xlabel('Component: {0}'.format(0))
ax.set_ylabel('Component: {0}'.format(1))
ax.set_zlabel('Component: {0}'.format(2))
x_size = (max(T[:,0]) - min(T[:,0])) * 0.08
y_size = (max(T[:,1]) - min(T[:,1])) * 0.08
z_size = (max(T[:,2]) - min(T[:,2])) * 0.08
# It also plots the full scatter:
ax.scatter(T[:,0],T[:,1],T[:,2], marker='.', alpha=0.65)
```
